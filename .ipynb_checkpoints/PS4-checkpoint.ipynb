### 2. Build a classifier using an SVM to identify whether a comment results to a recommendation or not.

```python
df["comments"] = [entry.lower() for entry in df["comments"]]
df.head(10)
```

```python
import nltk
nltk.download('punkt')
```

```python
#Tokenization : In this each entry in the comments will be broken into a set of words
df["comments"]= [word_tokenize(entry) for entry in df["comments"]]
```

```python
nltk.download('wordnet')
```

```python
# WordNetLemmatizer requires Pos tags to understand if the word is noun or verb or adjective etc. By default it is set to Noun
tag_map = defaultdict(lambda : wn.NOUN)
tag_map['J'] = wn.ADJ
tag_map['V'] = wn.VERB
tag_map['R'] = wn.ADV
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
nltk.download('stopwords')
```

```python
for index,entry in enumerate(df["comments"]):
    # Declaring Empty List to store the words that follow the rules for this step
    Final_words = []
    # Initializing WordNetLemmatizer()
    word_Lemmatized = WordNetLemmatizer()
    # pos_tag function below will provide the 'tag' i.e if the word is Noun(N) or Verb(V) or something else.
    for word, tag in pos_tag(entry):
        # Below condition is to check for Stop words and consider only alphabets
        if word not in stopwords.words('english') and word.isalpha():
            word_Final = word_Lemmatized.lemmatize(word,tag_map[tag[0]])
            Final_words.append(word_Final)
    # The final processed set of words for each iteration will be stored in 'text_final'
    df.loc[index,"text_final"] = str(Final_words)
```

```python
df.head()
```

```python
Train_X, Test_X, Train_Y, Test_Y = train_test_split(df["text_final"],df["recommended"],test_size=0.3)
```

```python
Encoder = LabelEncoder()
Train_Y = Encoder.fit_transform(Train_Y)
Test_Y = Encoder.fit_transform(Test_Y)
```

```python
Tfidf_vect = TfidfVectorizer(max_features=5000)
Tfidf_vect.fit(df["text_final"])
Train_X_Tfidf = Tfidf_vect.transform(Train_X)
Test_X_Tfidf = Tfidf_vect.transform(Test_X)
```

```python
# print(Tfidf_vect.vocabulary_)
```

```python
print(Train_X_Tfidf)
```

### Using Linear Kernel

```python
# fit the training dataset on the classifier
SVM = SVC(kernel='linear')
SVM.fit(Train_X_Tfidf,Train_Y)

# predict the labels on validation dataset
predictions_SVM = SVM.predict(Test_X_Tfidf)

# Get the accuracy
scores = cross_val_score(SVM, Test_X_Tfidf, Test_Y, cv=5)
acc = scores.mean()
print(f"SVM Accuracy Score: {acc*100}%")

print(metrics.classification_report(Test_Y, predictions_SVM))
```

### Using Poly Kernel

```python
# fit the training dataset on the classifier
SVM = SVC(kernel='poly')
SVM.fit(Train_X_Tfidf,Train_Y)

# predict the labels on validation dataset
predictions_SVM = SVM.predict(Test_X_Tfidf)

# Use accuracy_score function to get the accuracy
scores = cross_val_score(SVM, Test_X_Tfidf, Test_Y, cv=5)
acc = scores.mean()
print(f"SVM Accuracy Score: {acc*100}%")

print(metrics.classification_report(Test_Y, predictions_SVM))
```

### Using RBF Kernel

```python
# fit the training dataset on the classifier
SVM = SVC(kernel='rbf')
SVM.fit(Train_X_Tfidf,Train_Y)

# predict the labels on validation dataset
predictions_SVM = SVM.predict(Test_X_Tfidf)

# Use accuracy_score function to get the accuracy
scores = cross_val_score(SVM, Test_X_Tfidf, Test_Y, cv=5)
acc = scores.mean()
print(f"SVM Accuracy Score: {acc*100}%")

print(metrics.classification_report(Test_Y, predictions_SVM))
```

### Using Sigmoid Kernel

```python
# fit the training dataset on the classifier
SVM = SVC(kernel='sigmoid')
SVM.fit(Train_X_Tfidf,Train_Y)

# predict the labels on validation dataset
predictions_SVM = SVM.predict(Test_X_Tfidf)

# Use accuracy_score function to get the accuracy
scores = cross_val_score(SVM, Test_X_Tfidf, Test_Y, cv=5)
acc = scores.mean()
print(f"SVM Accuracy Score: {acc*100}%")

print(metrics.classification_report(Test_Y, predictions_SVM))
```

After comparison of accuracy scores from SVM's with different kernels, it was observed that all models resulted to close accuracies to each other, with the "sigmoid" kernel garnering the lowest cross-validation accuracy score and the "rbf" kernel garnering the highest cross-validation score.

### 4. Identify the top words for the pros and cons columns.

```python
df["pros"] = [entry.lower() for entry in df["pros"]]
df["cons"] = [entry.lower() for entry in df["cons"]]
df["comments"] = [entry.lower() for entry in df["comments"]]
df.head(10)
```

```python
pros_text = df["pros"].str.cat(sep=' ')
cons_text = df["cons"].str.cat(sep=' ')
comments_text = df["comments"].str.cat(sep=' ')
df_list = [pros_text, cons_text, comments_text]
```

```python
# instantiate the vectorizer object
vectorizer = CountVectorizer(stop_words="english")

# convert the documents into a document-term matrix
bag_of_words = vectorizer.fit_transform(df_list)
```

```python
def bag2df(bag, feat_names):
    
    # create an index for each row
    doc_names = ['Doc{:d}'.format(idx) for idx, _ in enumerate(bag)]
    df = pd.DataFrame(data=bag.toarray(), index=doc_names,
                      columns=feat_names)
    return(df)
```

```python
# retrieve the terms found in the corpora
tokens = vectorizer.get_feature_names()

# create a dataframe from the matrix
bag2df(bag_of_words, tokens)
```

```python
tfidf_vec = TfidfVectorizer(stop_words="english")
transformed = tfidf_vec.fit_transform(df_list)
```

```python
index_value={i[1]:i[0] for i in tfidf_vec.vocabulary_.items()}
fully_indexed = []
for row in transformed:
    fully_indexed.append({index_value[column]:value for (column,value) in zip(row.indices,row.data)})
```

```python
df_pros = pd.DataFrame(list(fully_indexed[0].items()),columns = ['word','weighted_value']) 
df_pros.head()
```

### Top 10 Words for the "Pros" Column

```python
df_pros.sort_values(by = 'weighted_value', ascending = False, inplace = True)
df_pros.head(10)
```

```python
freq_dict = {}
for word, value in fully_indexed[0].items():
    freq_dict[word] = value

wordcloud = WordCloud(width = 600, height = 600,
                background_color ='white',
                max_words = 50,
                min_font_size = 10).generate_from_frequencies(frequencies = freq_dict)

# plot the WordCloud image                       
plt.figure(figsize = (5, 5), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)
  
plt.show()
```
